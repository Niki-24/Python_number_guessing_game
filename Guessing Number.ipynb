## Features:
- Computer selects a number (e.g., 7)
User keeps guessing
**Give hints:**
- **Too high**
- **Too low**
- **Stop when correct**
### Skills Used:
- while loop
- if-else
- Count attempts

```python
secret_value = 7
count = 3
while count>0:
    num = int(input("Enter Your Secret Value:"))

    if num<0:
        print("Value less than 0 not allowed")
    elif num==secret_value:
        print("Congratulation You Win !!")
        break 
    elif num<secret_value:
        print("Too low")
    elif num>secret_value:
        print("Too high")
    else:
        print("Your value are Out of range!!")
    count-=1
    print(f"Attempt left:{count}")
if count==0:
    print("Game Over!! You lost") 
```
